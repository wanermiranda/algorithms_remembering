````python
import numpy as np 
from typing import Tuple
import random 

class Knight:
    MOVES = [(2, 1), (1, 2), (-1, 2), (-2, 1), (-2, -1), (-1, -2), (1, -2), (2, -1)]
            #[(2, 1), (1, 2), (-1, 2), (-2, 1), (-2, -1), (-1, -2), (1, -2), (2, -1)]

    def __init__(self, N=8, pos=(0,0)):
        self._board = np.ones((N,N), np.int) * -1
        self._board[pos[0], pos[1]] = 0 # starting position

        self._longest_time = 1
    
    def is_valid(self, future_pos):
        shp = self._board.shape    
        try:
            return  future_pos is not None and \
                    future_pos[0] < shp[0] and future_pos[1] < shp[1]  and \
                    future_pos[0] >= 0     and future_pos[1] >= 0      and \
                    self._board[future_pos[0], future_pos[1]] == -1
        except:
            print(shp, future_pos)
            raise Exception
    
    def mark_seen(self, pos:Tuple[int,int], time):
        self._pos = pos
        self._board[pos[0], pos[1]] = time

        if time >= self._longest_time:
            self._longest_time = time
            print(f"Pos {self._pos} at time {time}")
            print(self._board)
    
    def undo_move(self, pos):
        self._board[pos[0], pos[1]] = -1
    

    def solve_all_cells(self, pos=(0,0), time=1):
        if time == self._board.size:
            return True
        
        for x,y in Knight.MOVES:
            future_pos = (pos[0] + x, pos[1] + y)
            if self.is_valid(future_pos):
                
                self.mark_seen(future_pos, time)
                
                if self.solve_all_cells(future_pos, time+1):
                    return True
                
                self.undo_move(future_pos)
            
        return False
      
    def solve_all_cells_simpler(self, board, pos=(0,0), time=1):
        
        if time == board.size:
            return True
        
        for x,y in Knight.MOVES:
            future_pos = (pos[0] + x, pos[1] + y)
            if self.is_valid(future_pos):
                
                board[future_pos[0], future_pos[1]] = time
                
                if self.solve_all_cells_2(board, future_pos, time+1):
                    return True
                
                board[future_pos[0], future_pos[1]] = -1
                
        return False
    
    @staticmethod
    def random_move():
        return random.choice(Knight.MOVES)
````
